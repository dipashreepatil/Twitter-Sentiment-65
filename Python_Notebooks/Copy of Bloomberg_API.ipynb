```
# Write the data to json file
# Creating a list of lists of required parameters of news
news_list = []
with open("bloomberg_json.json","w", encoding = 'utf8') as file:
  for news in news_data:
    json.dump(news,file)
    file.write('\n')

    lst = []
    lst.extend((news['id'],news['internalID'],news['title'],news['abstract'],news['summary'],news['autoGeneratedSummary'],news['published']))
    news_list.append(lst)
```

```
#converting list of lists to csv file

df = pd.DataFrame(news_list,columns=['ID','InternalID','Title','Abstract','Summary','AutoGenerated Summary','Published At'])
df.to_csv('/content/drive/MyDrive/B-Tech Project 20-21 Group 7/Datasets/NLP datasets/DIPU Analysis Data/bloomberg_news.csv')
```
